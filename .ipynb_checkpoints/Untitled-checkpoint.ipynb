```python
import csv
import requests
from scrapy.http import TextResponse 
import pandas as pd
import pandasql as ps
import matplotlib.pyplot as plt


#Question 1
#get data form Data source in txt format , pick any 5 random days.
#combine them into one dataset and create a pandas DataFrame with specific structure

# 5 random days 
dates = [
       "2014-06-30", 
       "2014-06-29", 
       "2014-06-28",
       "2014-06-27", 
       "2014-06-26"
       ]

#remove old data if exit or create new file and add the headers     
with open('text.csv', 'w', newline='\n', encoding='utf-8') as file:
      writer = csv.writer(file)
      writer.writerow(
          ["CHAT_DAY", "CHAT_HOUR", "CHAT_MINUTE","CHAT_SECOND","USERNAME","CHAT_MESSAGE"]
          )  

def addDataToCSV(lines):
    with open('text.csv', 'a', newline='\n', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerows([line.split("!*|#%^") for line in lines])

def  get_txt(url,date):
    # Make a request to the URL and get the response
    response = requests.get(url)

    # Use Scrapy's TextResponse to parse the HTML content
    text_response = TextResponse(url=response.url, body=response.text, encoding='utf-8')

    # Get the text from the response
    text = text_response.text 

    # Split the text into lines -- default \n
    lines = text.split("\n") 

    #Add date then replace the first 2 spaces and : in each line with specific string that
    #will used to separate column to get CSV 
    data = []
    for line in lines:
         if len(line)!=0 :  
           line= date+"!*|#%^"+line.replace(":", "!*|#%^", 2)
           data.append( line.replace(" ", "!*|#%^", 2))  
    
    # Write the modified lines to a CSV file
    addDataToCSV(data) 
 
for date in dates:
  url="https://chatlogs.planetrdf.com/swig/"+date+".txt"
  get_txt(url,date) 
 
df = pd.read_csv('text.csv')




#Question 2a 
#Top 3 users by the most sent messages

mostSentMessages = '''
SELECT USERNAME, COUNT(CHAT_MESSAGE) as MESSAGE_COUNT
FROM df
GROUP BY USERNAME
ORDER BY MESSAGE_COUNT DESC
LIMIT 3
'''

result = ps.sqldf(mostSentMessages, locals())
print("Top 3 users by the most sent messages")
print(result)




#Question 2b
#Count of distinct users and messages sent per hour, sorted in ascending order by the hour
#plot this graph, alongside the mean messages sent per user per hour

messagesSentPerHour = '''
SELECT CHAT_HOUR, COUNT(DISTINCT USERNAME) as USER_COUNT, COUNT(CHAT_MESSAGE) as MESSAGE_COUNT
FROM df
GROUP BY CHAT_HOUR
ORDER BY CHAT_HOUR
'''
print("\nCount of distinct users and messages sent per hour")
result = ps.sqldf(messagesSentPerHour, locals())
print(result)


fig, ax1 = plt.subplots()

# Plot the bar chart for messages sent per hour
ax1.bar(result['CHAT_HOUR'], result['MESSAGE_COUNT'], alpha=0.5, align='center', color='blue', label='Messages sent')
ax1.set_xlabel('Hour')
ax1.set_ylabel('Messages sent', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
 



#Question 4
#Send the last-sent message of the username with the most messages sent to  API endpoint
#Print the value of the `id` key in the response payload.


# Get the username with the most messages sent
query = '''
SELECT USERNAME, COUNT(CHAT_MESSAGE) as MESSAGE_COUNT
FROM df
GROUP BY USERNAME
ORDER BY MESSAGE_COUNT DESC
LIMIT 1
'''

result = ps.sqldf(query, locals())
username = result['USERNAME'][0]

# Get the last message sent by the username with the most messages sent
query = f'''
SELECT CHAT_MESSAGE
FROM df
WHERE USERNAME = '{username}'
ORDER BY CHAT_DAY DESC, CHAT_HOUR DESC, CHAT_MINUTE DESC, CHAT_SECOND DESC
LIMIT 1
'''

result = ps.sqldf(query, locals())
last_message = result['CHAT_MESSAGE'][0]

print(f"\nThe Last Message is: {last_message }\n")


endpoint_url = 'https://63da8f4c19fffcd620cc7ea3.mockapi.io/coding-test/chatstats' 
data = {'username': username, 'chat_message': last_message}

response = requests.post(endpoint_url, data )
 
if response.status_code == 201:
    print('Successfully sent the message to API endpoint ') 
    print('id: ', response.json()['id'])
    
else:
    print(f'Failed to send the message. Response: {response.text}')
```
